```python
sns.set_theme(style="whitegrid")

for file_name in parquet_files:
    full_file_name = "/content/drive/MyDrive/datasets/" + file_name

    print(f"\n{'='*60}")
    print(f"üìÇ ANALIZA ZBIORU: {file_name}")
    print(f"{'='*60}")

    df = pd.read_parquet(full_file_name)

    print(f"Liczba wierszy: {len(df)}")
    print(f"Liczba kolumn: {len(df.columns)}")
    print("\nTypy danych:")
    print(df.dtypes)

    print("\nüîç BrakujƒÖce warto≈õci (Nulle):")
    null_counts = df.isnull().sum()
    if null_counts.sum() == 0:
        print("Brak pustych warto≈õci! ‚úÖ")
    else:
        print(null_counts[null_counts > 0])

    print("\nüìä Statystyki opisowe (Describe):")
    display(df.describe())

    print("\nüëÄ PodglƒÖd pierwszych wierszy:")
    display(df.head(3))

    if 'graph_data' in df.columns:
        print("\nüï∏ Przyk≈Çadowe dane grafowe (pierwszy rekord):")
        sample_graph = df.iloc[0]['graph_data']
        print(sample_graph)

    print(f"\nüß™ Przyk≈Çadowe czƒÖsteczki z {file_name}:")
    if 'input_smiles' in df.columns:
        smiles_list = df['input_smiles'].head(6).tolist()
        pIC50_list = df['target_pIC50'].head(6).tolist()

        mols = [Chem.MolFromSmiles(s) for s in smiles_list]

        valid_mols = [m for m in mols if m is not None]

        img = Draw.MolsToGridImage(
            valid_mols,
            molsPerRow=3,
            subImgSize=(300, 300),
            legends=[f"pIC50: {x:.2f}" for x in pIC50_list]
        )
        display(img)

    if 'target_pIC50' in df.columns:
        print(f"\nüìà Rozk≈Çad target_pIC50:")
        plt.figure(figsize=(10, 5))

        sns.histplot(data=df, x='target_pIC50', kde=True, bins=30, color='teal')

        plt.title(f'Rozk≈Çad warto≈õci pIC50 - {file_name}')
        plt.xlabel('pIC50 (logarytmiczna si≈Ça dzia≈Çania)')
        plt.ylabel('Liczba czƒÖsteczek')
        plt.axvline(df['target_pIC50'].mean(), color='red', linestyle='--', label=f'≈örednia: {df["target_pIC50"].mean():.2f}')
        plt.legend()
        plt.show()

    print("\n" + "-"*60 + "\n")
```

```python
for file_name in parquet_files:
    full_file_name = "/content/drive/MyDrive/datasets/" + file_name
    if not os.path.exists(full_file_name):
        print(f"\n‚ùå PLIK NIE ISTNIEJE: {full_file_name} (pomijam)")
        continue
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from rdkit import Chem
from rdkit.Chem import Draw
from IPython.display import display

sns.set_theme(style="whitegrid")


for file_name in parquet_files:
    full_file_name = "/content/drive/MyDrive/datasets/after_eda/" + file_name

    try:
        df = pd.read_parquet(full_file_name)
    except Exception as e:
        print(f"Nie uda≈Ço siƒô wczytaƒá pliku {file_name}: {e}")
        continue

    print(f"\n{'='*60}")
    print(f"üìÇ ANALIZA ZBIORU: {file_name}")
    print(f"{'='*60}")

    print(f"Liczba wierszy: {len(df)}")
    print(f"Liczba kolumn: {len(df.columns)}")
    print(f"Dostƒôpne kolumny: {list(df.columns)}")

    smiles_col = None
    for candidate in ['canonical_smiles', 'input_smiles', 'smiles']:
        if candidate in df.columns:
            smiles_col = candidate
            break

    target_col = None
    for candidate in ['target', 'pchembl_value', 'target_pIC50', 'label']:
        if candidate in df.columns:
            target_col = candidate
            break

    print(f"\nU≈ºywam kolumny SMILES: {smiles_col}")
    print(f"U≈ºywam kolumny Target: {target_col}")

    print("\nüìä Statystyki opisowe:")
    display(df.describe())

    if 'graph_data' in df.columns:
        print("\nüï∏ Przyk≈Çadowe dane grafowe (pierwszy rekord):")
        print(str(df.iloc[0]['graph_data'])[:200] + "...")

    print(f"\nüß™ Przyk≈Çadowe czƒÖsteczki z {file_name}:")
    if smiles_col and target_col:
        subset = df.head(6)
        smiles_list = subset[smiles_col].tolist()
        pIC50_list = subset[target_col].tolist()

        mols = [Chem.MolFromSmiles(s) for s in smiles_list]

        valid_mols = []
        legends = []
        for mol, val in zip(mols, pIC50_list):
            if mol:
                valid_mols.append(mol)
                legends.append(f"pIC50: {val:.2f}")

        if valid_mols:
            img = Draw.MolsToGridImage(
                valid_mols,
                molsPerRow=3,
                subImgSize=(300, 300),
                legends=legends
            )
            display(img)
        else:
            print("Nie uda≈Ço siƒô wygenerowaƒá obrazk√≥w (b≈Çƒôdne SMILES).")
    else:
        print("‚ùå Nie znaleziono kolumny SMILES lub Target do wizualizacji.")

    if target_col:
        print(f"\nüìà Rozk≈Çad {target_col}:")
        plt.figure(figsize=(10, 5))

        df[target_col] = pd.to_numeric(df[target_col], errors='coerce')

        sns.histplot(data=df, x=target_col, kde=True, bins=30, color='teal')

        plt.title(f'Rozk≈Çad warto≈õci pIC50 - {file_name}')
        plt.xlabel('pIC50 (logarytmiczna si≈Ça dzia≈Çania)')
        plt.ylabel('Liczba czƒÖsteczek')

        mean_val = df[target_col].mean()
        plt.axvline(mean_val, color='red', linestyle='--', label=f'≈örednia: {mean_val:.2f}')
        plt.legend()
        plt.show()

    print("\n" + "-"*60 + "\n")
```
